### **Importando as bibliotecas**

Por enquanto, só será utilizado o pandas pra manipular os dados.

```python
import pandas as pd
```

### **Transformando o arquivo .csv em um dataframe pandas**

```python
tmdb_df = pd.read_csv('tmdb.csv')
tmdb_df.head()
```

```python
tmdb_df.info()
```

Aqui, é possível ver que há 14999 linhas, mas há algumas colunas com valores nulos.

### **Retirar os filmes que ainda não foram lançados**

Os filmes que não foram lançados, ou que estão como especulados, não servem para fazer o machine learning. Então, todas as linhas em que a coluna `status` esteja diferente de `Released` serão excluídas.

```python
tmdb = tmdb_df[tmdb_df['status'] == 'Released']
tmdb.head()
```

```python
tmdb.info()
```

### **Remover as colunas que não serão usadas**

Algumas colunas não serão o foco do projeto, ou serão muito trabalhosas para tratar.

```python
tmdb = tmdb.drop(columns=['overview', 'poster_path', 'backdrop_path', 'cast', 'crew', 'tagline', 'status'])
tmdb.head()
```

```python
tmdb.info()
```

### **Conferir quantos valores nulos existem em cada coluna**

```python
print(tmdb.isna().sum())
```

Todas as colunas que possuem dados nulos são importantes para o machine learning e os dados não podem ser inferidos através de outros, então precisam ser excluídos.

### **Remover todos os dados nulos**

```python
tmdb_clean_df = tmdb.dropna()
tmdb_clean_df.tail()
```

```python
print(tmdb_clean_df.isna().sum())
```

Agora, o dataframe não tem mais dados vazios.

### **Resetar o índice das linhas**

Como algumas linhas foram apagadas, o índice das linhas está com algumas lacunas que precisam ser ajeitadas.

```python
tmdb_clean_df.reset_index(drop=True, inplace=True)
tmdb_clean_df.tail()
```

```python
tmdb_clean_df.info()
```

Após a limpeza, o dataframe que tinha 14999 linhas agora tem 14194, ou seja, foram excluídas 805 linhas.

### **Identificar os valores únicos das colunas de: gênero, língua original, empresa da produtora, país da produtora**

Esses dados podem ser transformados em números para trabalhar com os dados de uma forma mais objetiva.

```python
tmdb_clean_df['genres'].drop_duplicates()
```

> Existem 2839 diferentes combinações de gêneros de filmes. Seria interessante pensar em algum jeito de catalogar cada gênero de forma única (ex.: horror = 0, thriller = 1, mystery = 2) e depois catalogar o gênero de cada filme de acordo com esses números (ex.: linha índice 0 -> 0, 1, 2).

```python
tmdb_clean_df['original_language'].drop_duplicates()
```

> Existem 64 tipos difentes de idiomas no dataframe. Essa coluna está mais fácil de traduzir as línguas para números.

```python
tmdb_clean_df['production_companies'].drop_duplicates()
```

> Existem 11964 diferentes combinações de produtoras de filmes. É um número muito que talvez seja muito trabalhoso de tratar esses dados para eles ficarem mais  objetivos.

```python
tmdb_clean_df['production_countries'].drop_duplicates()
```

> Existem 1175 diferentes combinações de países de produtoras de filmes. É u número mais simples d se trabalhar do que o nome das empresas e seria interessante pensar em algum jeito de catalogar cada país de forma única (ex.: france = 1, united kingdom = 2) e depois catalogar o país de cada produtora de acordo com esses números (ex.: linha índice 3 -> 1, 2).

### **Atribuir um id a cada idioma**

```python
idioma_original = tmdb_clean_df['original_language'].drop_duplicates().reset_index(drop=True)
language_id = {lang: idx for idx, lang in enumerate(idioma_original)}

tmdb_clean_df['original_language_id'] = tmdb_clean_df['original_language'].map(language_id)
tmdb_clean_df
```

```python
tmdb_clean_df['original_language_id'].drop_duplicates()
```
